What is the distribution of skip rate across sessions?

```python
# Plots distribution of skip rate across sessions
plt.figure()
plt.hist(session_logs_df.groupby('session_id')['skip_2'].mean())
plt.axvline(session_logs_df['skip_2'].mean(), color='red', label=f"Mean skip rate: {session_logs_df['skip_2'].mean():.2f}")
plt.title("Distribution of Skip Rate Across Sessions")
plt.legend()
plt.xlabel('Skip Rate')
plt.ylabel('Count')
plt.show()
```

Based on the figure above it would appear that skip rate across sessions is somewhat uniform. However there appears to be more sessions with skip rates above 50% than below 50%.

---

How does skip rate differ based on time of day?

```python
# Plots skip rate vs. time of day
plt.figure()
session_logs_df.groupby('hour_of_day').mean()['skip_2'].plot()
plt.xticks(range(0,24))
plt.title('Skip Rate based on Time of Day')
plt.xlabel('Time (24 h)')
plt.ylabel('Skip Rate')
plt.show()
```

Based on the figure above, it appears that skip rate peaks at 2:00 AM. However, it should be noted that this peak in skip rate is only slightly higher than the other local peaks in skip rate found in the afternoon. It is interesting to note that skip rates appear to be lowest around 6:00 AM and then again at 9:00 AM. We might consider that users might be busy getting ready to start the day. Or might be commuting to work and therefore less likely to skip songs to focus on arriving to their destination safely. Same might be said for the afternoon/evening commute. Although there does seem to be some oscillations in skip rate through the course of a day which is also interesting to consider.

---

Another way to visualize the findings from above. **This one was originally a bit confusing when plotted on a linear scale**, but I fixed it by plotting on a log-scale.

```python
# Plots total songs played in the hour, and the total skips in that hour
plt.figure()
session_logs_df.groupby('hour_of_day').sum()['skip_2'].plot(label='Total songs skipped')
session_logs_df['hour_of_day'].value_counts().sort_index().plot(label='Total songs played')
plt.yscale('log')
plt.legend()
plt.xticks(range(0,24))
plt.title('Proportion of Skipped Songs vs. Total Songs Played, over Time')
plt.xlabel('Time (24 h)')
plt.ylabel('Count')
plt.show()
```

This is another way to present skip rate over time. We can see that the gap between songs played and songs skipped appears to be constant as we move through time. This suggests that skip rate is nearly constant throughout the course of the day. Which in general tells us that perhaps there might be a only a slight difference in skip rate based on the time of the day.

Does skip rate differ based on the type of playlist the user is listening to?

```python
# Gets the skip rate across different 'context_type', AKA playlists
round(session_logs_df.groupby('context_type')['skip_2'].mean().sort_values(ascending=False), 3)
```

In contrast, in top charts-type playlists, the listener might be looking for some popular music to listen to but aren't interested in every song in the playlist, thus they'd be more inclined to skip certain songs.

Does skip rate differ based on `'session_position'?

```python
# Plots skip rate across session position
plt.figure()
plt.plot(session_logs_df.groupby('session_position')['skip_2'].mean())
plt.xticks(range(1,21))
plt.axhline(session_logs_df['skip_2'].mean(), color='red', label='Mean skip rate across all sessions')
plt.title('Skip Rate based on Song Order')
plt.xlabel('Track Number')
plt.ylabel('Skip Rate')
plt.legend()
plt.show()
```

Skip rate appears to track consistently beyond the third track in a given listening session. However, it's interesting that the mean skip rate for the first track in a listening session is considerably lower than the rest of the songs in the session. I would imagine that users are not typically looking to skip the first track in a session by default. This might help inform us on predicting track skips for the first song in a given session. I.E. It might be useful to know whether or not the song is the first in the sequence or not. Beyond that it might be a bit difficult to predict skip solely based on where the song fell in the sequence.

---

Let's get a bit more granular and see how users because across a session based on the playlist they listened to.

```python
# Gets a list of the different context_types (playlist types)
context_types = ([col for col in session_logs_df['context_type'].unique()])

# Loops through and plots average skip rate based on track position and 
# context_type
for context in context_types:
    df2 = session_logs_df[session_logs_df['context_type'] == context]
    plt.figure()
    plt.plot(df2.groupby('session_position')['skip_2'].mean())
    plt.xticks(range(1,21))
    plt.axhline(session_logs_df['skip_2'].mean(), color='red', label='Mean skip rate across all sessions')
    plt.title(context)
    plt.xlabel('Track Number')
    plt.ylabel('Skip Rate')
    plt.legend()
    plt.show()
```

Some interesting things to note:
- Both radio and chart-type playlists have above average skip rates across the track position
    - With chart-type playlists varying quite a bit across track positions
    - I wonder if some of the users are skipping or fishing for certain tracks in the playlist
    - For instance, just searching for some songs they know they'd like to hear, but aren't interested in using a search function for it
    - Would be nice to know what platform the users listened on
    - I would imagine that it's much easier to hit the skip button than search for a song
    - Moreover, we see that skip rate in radio-type playlists is typical higher than average
    - However, in contrast to chart-type playlists, the variance across track position appears to be less
- In the `'personalized_playlist'` plot, there appears to be quite a bit of variance in skip rate across the listening session, however, it appears the skip rate is still below the average across the track positions
- Something important to note is that track skip rate is lowest during the first track of the session
- In addition, the variance track position indiscriminate of `'context_type'` appears to be smoothed out by `user_collection, catalog, and editorial_playlist `

**In conclusion, it appears that track position might matter in predicting skip behaviour if the song is one of the first few to appear in the beginning of the session.**

Is the way a song start associated with whether it is skipped or not?


```python
# Gets a list of the different context_types (playlist types)
reason_start_types = ([reason for reason in session_logs_df['hist_user_behavior_reason_start'].unique()])

# Loops through and plots average skip rate based on track position and 
# context_type
for reason in reason_start_types:
    df3 = session_logs_df[session_logs_df['hist_user_behavior_reason_start'] == reason]
    plt.figure()
    plt.plot(df3.groupby('session_position')['skip_2'].mean())
    plt.xticks(range(1,21))
    plt.axhline(session_logs_df['skip_2'].mean(), color='red', label='Mean skip rate across all sessions')
    plt.title(reason)
    plt.xlabel('Track Number')
    plt.ylabel('Skip Rate')
    plt.legend()
    plt.show()
```

Above are just some quick plots to see how skip rate differs based on how the user arrived at the current song, and how this differs across track position. I will touch on a couple of these plots that I think show quite interesting patterns.

```python
df4 = session_logs_df[session_logs_df['hist_user_behavior_reason_start'] == 'trackdone']

plt.figure()
plt.plot(df4.groupby('session_position')['skip_2'].mean())
plt.xticks(range(1,21))
plt.axhline(session_logs_df['skip_2'].mean(), color='red', label='Mean skip rate across all sessions')
plt.title("Skip Rate of Current Track if Previous Track was Played Through Entirely")
plt.xlabel('Track Number')
plt.ylabel('Skip Rate')
plt.legend()
plt.show()
```

```python
df5 = session_logs_df[session_logs_df['hist_user_behavior_reason_start'] == 'fwdbtn']

plt.figure()
plt.plot(df5.groupby('session_position')['skip_2'].mean())
plt.xticks(range(1,21))
plt.axhline(session_logs_df['skip_2'].mean(), color='red', label='Mean skip rate across all sessions')
plt.title("Skip Rate of Current Track if User Pressed Forward Button on Previous Track")
plt.xlabel('Track Number')
plt.ylabel('Skip Rate')
plt.legend()
plt.show()
```

The figures above tell us above the skip rate at each position during the session based on whether the user let the track end, or if they pressed the forward button, to arrive at the current song. These findings seem to suggest that a user might be less likely to skip if they listened to the previous in its entirely. In constrast, if a used the forward button to get to the current song. They might be more likely to skip the current song. 

This might be something good to keep in mind as I think about how I might construct my machine learning model. That previous user interaction behaviour might help me predict whether a user will skip the current track.

# Taking a look at the correlation between variables

```python
# Plots heatmap of correlation between variables
mask = np.triu(np.ones_like(main_df.corr()))

plt.figure(figsize=(20,10))
sns.heatmap(main_df.corr(), center = 0, vmin=-1, vmax=1, mask=mask, cmap='Spectral_r')
plt.show()
```

Above is a (rather large) heatmap for correlations between each variable in the dataset. However, as I suspected in the previous notebook while looking at the distributions for `'beat_strength, 'bounciness', and 'danceability'`. The distributions follow very similar shapes. In this case there are strong positive correlations between them. Meaning if the `beat_strength` of a song is known, it's likely I can predict the song's `'bounciness or danceability'` value. The same can be said for `'organism' and 'mechanism'`. Which appear to be opposites of each other, which in the heatmap is reflected as having a strong negative correlation. Meaning that if a song scores high in 'organism' it like scores low in 'mechanism'. Finally, pause behaviours appear to show strong negative correlations between themselves, similar to `'organism'` and `'mechanism'`. 

It should be noted that the correlation between the independent variables (or multicollinearity) mentioned above might pose a risk to our interpretability of certain models. For example, if `'beat_strength, 'bounciness', and 'danceability'` are all included in the same logistic regression model. The model runs the risk of be unable to weigh the coefficients properly across these variables. Accuracy is not expected to faulter, however my ability to interpret which coefficients are important to the model will.

It's interesting to note that `'skip_2'` our target variable does not appear to share any strong correlations with the other variables. Which seems to suggest that the features and interactions within a given observation is not strongly related with whether the song was skipped. Baseline modelling which these features as is might not yield the best results based on these findings. It might take a bit of extra work to transform the data in a way that is able to predict track skip.
